```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
```

```python
# 直接从列表画图
x = [1,2,3,4,5,6,7]
y = [65, 80, 75, 83, 77, 100, 98]
plt.plot(x,y)
plt.show()
```

```python
# 从NumPy画图
x = np.arange(0,7)
y = np.array([65, 80, 75, 83, 77, 100, 98])
plt.plot(x,y)
plt.show()
```

```python
# 从pandas画图
mydata = pd.DataFrame(
    {
        "Subject": ["军训", "思修", "史纲", "马原", "毛概", "体育1", "体育2"],
        "Grade": [65, 80, 75, 83, 77, 100, 98],
    }
)

df = pd.DataFrame(mydata)
df.plot.line()
plt.show()
```

```python
# 添加坐标信息，重新绘图
# plt.figure(figsize=(10,8),dpi=100)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin(x)")
plt.show()
```

```python
# 修改坐标轴刻度
plt.plot(x,y)
plt.ylim([0,700])
plt.show()
```

```python
# 画多条线
x = np.linspace(0, 20, 100)
y1 = np.sin(x)
y2 = np.sin(x / 2)

plt.figure(dpi=350)
plt.plot(x, y1,'r--^')
plt.plot(x, y2,'b:.')

# plt.plot(x,y1,'r--^')# 改变线形、颜色
# plt.plot(x,y2,'b:.')

plt.legend(["y1","y2"])# 图例
plt.title("正弦波")# 标题

plt.xlim([0,30])# x坐标范围
plt.ylim([-2,2])# y坐标范围
plt.xticks(np.linspace(0,20,3))# x坐标标签

plt.show()
```

```python
plt.rcParams["font.sans-serif"] = ["SimHei"]  # 用来正常显示中文标签
plt.rcParams["axes.unicode_minus"] = False  # 用来正常显示负号

plt.figure(dpi=100)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["y1","y2"])
plt.title("电压输出曲线")
plt.show()
```

```python
# 子图：subplot
plt.figure(dpi=100)
plt.subplot(221)  # 两行，两列，第一幅图
plt.plot(x, y1)

plt.subplot(222)  # 两行，两列，第二幅图
plt.plot(x, y2)

plt.subplot(223)  # 两行，两列，第三幅图
plt.plot(x, y1)
plt.plot(x, y2)

plt.subplot(224)  # 两行，两列，第四幅图
plt.plot(x, y2)
plt.plot(x, y1)

plt.show()
```
