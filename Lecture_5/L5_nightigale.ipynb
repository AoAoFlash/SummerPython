```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams["font.sans-serif"] = ["SimHei"]  # 用来正常显示中文标签
plt.rcParams["axes.unicode_minus"] = False  # 用来正常显示负号
```

```python
# 从“等长列表组成的字典”构造DataFrame
mydata = pd.DataFrame(
    {
        "Subject": ["军训", "思修", "史纲", "马原", "毛概", "体育1", "体育2"],
        "Grade": [65, 80, 75, 83, 77, 100, 98],
    }
)
mydata
```

```python
n_row = mydata.shape[0]
x = np.arange(0, 2 * np.pi, 2 * np.pi / n_row)
y = np.array(mydata["Grade"])

print(x)
print(y)
```

```python
# 折线图
fig = plt.figure()  # 创建新图窗
plt.plot(x, y)
plt.show()
```

```python
# 条形图
fig = plt.figure()  # 创建新图窗
plt.bar(x, y)
plt.show()
```

```python
# 南丁格尔玫瑰图
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True)  # 设置为极坐标，polar=True
plt.bar(x, y)
plt.show()
```

```python
# 增加细节
fig = plt.figure(dpi=200)  # 创建新图窗
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True)  # 设置为极坐标，polar=True

plt.xticks(x, labels=mydata["Subject"], size=14)  # 把x轴的坐标标签换成 mydata["Subject"]

plt.ylim(0, 100)  # 限制y轴的范围
plt.grid(alpha=0.2)  # 网格的透明度
plt.bar(x, y, color="#F94306", width=0.7, alpha=0.6)  # 颜色、宽度、透明度……
# plt.yticks([])# 隐藏y坐标

# 旋转y坐标的角度
ax.set_theta_direction(-1)  # -1：逆时针，1：顺时针
ax.set_theta_offset(np.pi / 2 - np.pi / n_row)
ax.set_rlabel_position(360 - 180 / n_row)  # 方法用于设置极径标签显示位置,参数为标签所要显示在的角度

plt.show()
# plt.savefig("南丁格尔玫瑰图.png")
# plt.close()
```
