```python
import matplotlib.pyplot as plt
```

```python
# 直接从列表画图
x = [1,2,3,4,5,6,7,8]
y = [a**2 for a in x]
plt.plot(x,y)
plt.show()
```

```python
# 从NumPy画图
import numpy as np 
x = np.arange(0,20,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
# 添加坐标信息，重新绘图
# plt.figure(figsize=(10,8),dpi=100)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin(x)")
plt.show()
```

```python
# 修改坐标轴刻度
plt.plot(x,y)
# plt.xlim([2,15])
plt.ylim([-0.8,0.8])

plt.xticks(np.linspace(0,20,3))

plt.show()
```

```python
# 画多条线
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.sin(x / 2)

plt.figure(dpi=100)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# 图例
plt.figure(dpi=100)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["y1","y2"])
plt.show()
```

```python
# 显示中文
plt.figure(dpi=100)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["y1","y2"])
plt.title("电压输出曲线")
plt.show()
```

```python
plt.rcParams["font.sans-serif"] = ["SimHei"]  # 用来正常显示中文标签
plt.rcParams["axes.unicode_minus"] = False  # 用来正常显示负号

plt.figure(dpi=100)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["y1","y2"])
plt.title("电压输出曲线")
plt.show()
```

```python
plt.figure(dpi=100)
plt.subplot(221)
plt.plot(x,y1)

plt.subplot(222)
plt.plot(x,y2)

plt.subplot(223)
plt.plot(x,y1)
plt.plot(x,y2)

plt.subplot(224)
plt.plot(x,y2)
plt.plot(x,y1)

plt.show()
```
